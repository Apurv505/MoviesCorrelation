```python
import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12, 8)
```

```python
df = pd.read_csv(r"C:\Users\Apurv\Downloads\movies.csv\movies.csv")
```

```python
df.head()
```

```python
for col in df.columns:
   
    pct_missing = np.mean(df[col].isnull())
    print('{}-{}%'.format(col, pct_missing))
```

```python
df = df.dropna()
```

```python
for col in df.columns: 
    
   print(df.isnull().any())
```

```python
df.dtypes
```

```python
df.loc[:, ('budget')] =  df.loc[:,('budget')].astype('int64')
df.loc[:, ('gross')] =    df.loc[:, ('gross')].astype('int64')
```

```python
df.head()
```

```python
df.sort_values(by ='gross',inplace=False, ascending=False)
```

```python
pd.set_option('display.max_rows', 100)
```

```python
df.drop_duplicates(inplace=True)
```

```python
plt.scatter(x=df['budget'], y=df['gross'])

plt.title('Budget vs Gross')
plt.xlabel('Gross earnings')
plt.ylabel('Budget')
plt.show()
```

```python
sns.regplot(x='budget', y='gross', data=df, scatter_kws={'color':'red'}, line_kws={'color':'cyan'})
```

```python
df.corr('pearson') #Pearson, kendall, spearman
```

```python
correlation_matrix = df.corr()
```

```python
sns.heatmap(correlation_matrix, annot=True, )
plt.title('Correlation Matrix')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
df_numerized = df
for col_name in df_numerized.columns:
    if (df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
df_numerized
```

```python
df
```

```python
correlation_matrix = df_numerized.corr()
```

```python
sns.heatmap(correlation_matrix, annot=True, )
plt.title('Correlation Matrix')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
correlation_mat = df_numerized.corr()
corr_pairs = correlation_mat.unstack()
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```
